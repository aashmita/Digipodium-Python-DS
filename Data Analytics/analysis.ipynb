```python
#Index based selection : iloc
df.head()  # top 5 rows
```

```python
df.iloc[2]  #to get the 3rd row of the dataframe
df.iloc[-1]  #negative indexing ->last row

# indexing starts from 0
```

```python
df.iloc[1:5]  #slicing->selecting continuos data
df.iloc[1:11:2]
```

```python
df.iloc[[3,5,7,11,17,2]]  # getting specific data
```

```python
df.iloc[1:6,1:6] # to get the rows from 1 to 5 and columns from 1 to 5
```

```python
df.iloc[[1,3,5],[1,3,5]] #to get the rows 1,3,5 and columns 1,3,5
```
