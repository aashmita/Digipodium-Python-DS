```python
import matplotlib.pyplot as plt
import seaborn as sns
```

From where this dataset came?

```python
df = sns.load_dataset('iris')
df
```

```python
# histogram -> to make graph from single column

# histogram -> ditributin of sepal length

plt.hist(df['sepal_length'],color='skyblue',edgecolor='black')
plt.title('Distribution of sepal length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

for line plot we do not need to enter plt.line or anything like plt.hist

```python
#line plot - petal length of first 50 samples
plt.plot(df['petal_length'][:50],color='orange') #this colon is slicing
plt.title('Petal Length of first 50 samples')
plt.xlabel('Sample Index')
plt.ylabel('Petal Length')
plt.show()
```

```python
#calculate mean values for each species
mean_value=df.groupby('species').mean()
```

```python
mean_value.plot(kind='bar',figsize=(5,5))
plt.title("Average Feature Values per Species")
plt.ylabel("Average Value")
plt.xticks(rotation=20) #xlabel rotation value
plt.show()
```

```python
# creating a figure and axis object
fig, ax = plt.subplots(figsize=(7,4))  #5,3

#setting the colour of figure and axes
fig.set_facecolor('white')
ax.set_facecolor('lightgreen')

#plotting the data
x=[1,2,3,4,5,6,7,8,9]
y=[3,6,2,4,7,9,8,2,5]

ax.plot(x,y,
        color='red',
        linewidth=1,
        linestyle='--',
        marker='o',
        markersize=4,
        markerfacecolor='white',
        markeredgecolor='black',
        markeredgewidth=2)


ax.set_title('Customised Line Chart', color='red', fontsize=18)
ax.set_xlabel("X-axis", color='green', fontsize=15)
ax.set_ylabel("Y-axis", color='green', fontsize=20)

#add text
ax.text(3,7,'Sample Text')
ax.axhline(y=5, color='black',alpha=0.2)
ax.axvline(x=5, color='black',alpha=0.2)

```
