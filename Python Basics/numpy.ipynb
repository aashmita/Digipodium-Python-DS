Core python - last topic -> numpy

- Python library used for numerical computations
- faster than list -> array > list
- more storage efficient

- -> numpy is a python library for numerical computing that provides fast and memeory-efficient arrays.
Compared to python lists, numpy array store same - type elements( in list we can store different data type) and perform mathematical operations faster.

```python
import numpy as np
a = np.array([1,2,3,4,5])
print(a)
```

```python
np.zeros(5)  # function -> gives zeroes and ones only
```

```python
np.ones(5)
```

```python
np.full(5,7)           # first index -> no of elements and second index ->value
```

```python
np.full((5,7),2)      # i-no of rows , j- no of columns , 2 - value
```

```python
np.arange(1,11,2) # starting index , last index , difference
```

```python
np.random.rand(3,3)  #random matrix within the range of 0 and 1
```

```python
np.eye(3)  # identity matrix of size 3*3
```

```python
np.random.randint(10,50,size=(3,3)) #random matrix within the range of 10 and 50   ->dtype=int32
```

```python
np.array([[1,2,3],[4,5,6]]) #2d array
```
