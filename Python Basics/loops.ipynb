```python
# break

fruits = ['apple', 'mango', 'banana', 'cherry', 'tomato', ' guava']
for x in fruits:
    if x=="tomato":
        break
    print(x)
```

```python
# continue -> skips that element and moves further

fruits = ['apple', 'mango', 'banana', 'cherry', 'tomato', 'guava']
for x in fruits:
    if x=="tomato":
        continue
    print(x)
```

```python
num=11
if num<=2:
    print("not a prime number")
else:
    for i in range(2,num):
        if num%i==0:
            print("not a prime number")
            break   
    else:
        print("prime number")
```

```python
end = int(input("Enter end range"))
prime_lst = []

for num in range(2, end+1):
    for i in range(2,num):
        if num%i == 0:
            break
    else:
        prime_lst.append(num)
print(prime_lst)

```
