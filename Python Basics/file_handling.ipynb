```python
file = open("data.txt","r")
print(file.name)  #name of the file
print(file.mode) # mode in which file is open
print(file.closed) # check if the file is closed or not
print(file.tell())  #current position of the file cursor
file.read(5)
print(file.tell())
file.seek(0) #move the cursor to the beginning of the file
```

Exception handling

```python
print("Opening existing file")
try:
    file = open("data.txt","r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not found.")

print("Opening random file")
try:
    file = open("dataa.txt","r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not found.")
```
