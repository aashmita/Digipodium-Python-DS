choice = input("enter a choice : \n a - check balance, b - deposit, c - withdraw, d - exit")

balance = 10000

while True:
    if choice == 'a':
        print("Your balance is ", balance)
    elif choice == 'b':
        amt = int(input("enter amount to deposit : "))
        balance = balance + amt
        print("Your new balance is : ",amt)
    elif choice == 'c':
        amt = int(input("Enter the amount to withdraw : "))
        balance = balance - amt
        print("Your new amount is : ",balance)
    elif choice == 'd':
        print("Thank you for using our service")
        break
    else:
        print("Invalid choice, please try again")
