## Conditional Statements - 25 May 2026

- Used for decision making statements
- can have multiple conditions
- kind of represent blocks
- one block is represented by curly braces {}
- function, class, loop
- Block is a set of statements


```python
# if - else block -> checking DL eligibility

age = int(input("Enter your age"))

if age>=18:
    print("Eligible for driving license")
else:
    print("Not eligible")
    
print(" Results are out")
```

```python
# WAP to check if a number is even
print(5%2==0) # returns if even -> true ; if odd -> false
```

```python
num = int(input("Enter a number"))

if num%2==0:
    print("Even")
else:
    print("False")
```

```python
marks = float(input("Enter marks"))

if marks<=100 and marks>=90:
    print("Grade A")
elif marks<90 and marks>=70:
    print("Grade B")
elif marks<70 and marks>=50:
    print("Grade C")
elif marks>100 or marks<0:
    print("Invalid Input")
else:
    print("Fail")
```

```python
# find largest from 3 numbers

a, b, c = 40, 80, 50

if a>b:
    if a>c:
        print(f'{a} is largest')

    else:
        print(f'{c} is largest')

else:
    if b>c:
        print(f'{b} is largest')
    else:
        print(f'{c} is largest')



```
