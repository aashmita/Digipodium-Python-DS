- calculate area of circle where radius is 5
- calculate area of square
- basic salary of an employee is 50,000. HRA is 5% of basic salary and DA is 10% of bs. Calculate gross salary
- take inpute from user and calculate simple interest.

```python
# area of circle
r = int(input("Enter radius"))
area = 3.14*r*r
print("area of circle is ",area)
```

```python
# area of square
a = int(input("enter side"))
area = a*a
print(f"area of square with side {a} is {area}")
```

```python
#basic salary = 50000
#HRA = 5% OF BASIC
#DA = 10% OF BASIC
# CALCULATE GROSS = SUM OF ALL

BS = 50000
HRA = (5*BS)/100
DA = (10*BS)/100
GROSS= BS+HRA+DA
print("gross salary is ",GROSS)
```

```python
# simple interest = (principle * rate * time)/100
P = int(input("Enter Principle amount"))
R = int(input("Enter rate"))
T = int(input("Enter time"))
SI = (P*R*T)/100
print(f"Simple Interest with principle {P} rate {R} and time {T} is {SI}")
```
